```python
from collections import defaultdict

class Solution:
    def invalidTransactions(self, transactions):
        
        if not transactions:
            return []
        
        split = [transaction.split(",") for transaction in transactions]
        res = set()
        
        name_to_records = defaultdict(list)
        for name, minutes, amount, city in split:
            name_to_records[name].append((minutes, amount, city))
        
        for name, records in name_to_records.items():
            
            records.sort(key=lambda x: int(x[0]))
            
            prev_minutes = records[0][0]
            prev_city = records[0][2]
            
            print(name, records)
            
            for i, (minutes, amount, city) in enumerate(records):
                if int(minutes)-int(prev_minutes) <= 60 and prev_city != city:
                    res.add(",".join([name, minutes, amount, city]))
                    res.add(",".join([name] + list(records[i-1])))
                elif int(amount) > 1000:
                    res.add(",".join([name, minutes, amount, city]))
                prev_minutes = minutes
                prev_city = city
                
        return list(res)
```
